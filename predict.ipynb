```python
plastic_no_map = {
  0: "PP",
  1: "PE",
  2: "PET",
  3: "PS"
}

color_name_map = {
  "white": 0,
  "gray": 1,
  "black": 2,
  "red": 3,
  "brown": 4,
  "yellow": 5,
  "green": 6,
  "blue": 7
}
```

```python
import numpy as np
import cv2
from tensorflow.keras.models import load_model
```

```python
def calc_luminance(img: np.ndarray) -> np.ndarray:
    hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
    h,l,s = cv2.split(hls)
    raveled = l.ravel()

    l_count = np.zeros(256, dtype=np.int32)
    for i in range(256):
        l_count[i] = raveled[raveled == i].shape[0]
    
    return l_count
```

```python
model = load_model("plastic-classification.h5")
```

```python
color_name = "blue"
img_850 = cv2.imread("PE_Blue_850nm.png")
img_940 = cv2.imread("PE_Blue_940nm.png")
```

```python
info = np.concatenate([
    [color_name_map[color_name]],
    calc_luminance(img_850),
    calc_luminance(img_940)
])

data = np.zeros((1,513), dtype=np.int32)
data[0] = info

plastic_no = np.argmax(model.predict(data))
```

```python
print("これは {} です".format(plastic_no_map[plastic_no]))
```
